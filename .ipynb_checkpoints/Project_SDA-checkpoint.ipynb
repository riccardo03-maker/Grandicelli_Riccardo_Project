# Step 1: LD from real data
## Data reading

```r
library(tidyverse)
```

```r
download.file(
  "https://ftp.1000genomes.ebi.ac.uk/vol1/ftp/release/20130502/integrated_call_samples_v3.20130502.ALL.panel",
  destfile = "integrated_call_samples_v3.20130502.ALL.panel"
)
```

```r
genes <- data.frame( #data frame with the genes
  gene  = c("HLA-DRB1","TP53","ABO"),
  chromosome   = c("6", "17", "9"),
  start = c(32579019, 7661882, 133233496), #start and stop from gnomAD dataset
  end   = c(32589816, 7676669, 133275264),
  stringsAsFactors = FALSE
)

populations<-c("EUR", "AFR", "EAS") #the three populations studied
cmd<-"plink"
```

```r
panel <- read.delim(
  "integrated_call_samples_v3.20130502.ALL.panel",
  stringsAsFactors = FALSE
)

write_keep <- function(superpop, out_file) {
  ids <- panel$sample[panel$super_pop == superpop]
  keep <- data.frame(FID = ids, IID = ids)
  write.table(keep, out_file,
              quote = FALSE, row.names = FALSE,
              col.names = FALSE, sep = "\t")
}

write_keep("EUR", "EUR.keep")
write_keep("AFR", "AFR.keep")
write_keep("EAS", "EAS.keep")
```

```r
dir.create("plink_out", showWarnings = FALSE) #create a new folder with the results if it does not exist
for (i in 1:3){
    file_path<-sprintf("vcf_files/ALL.chr%s.phase3_shapeit2_mvncall_integrated_v5a.20130502.genotypes.vcf.gz", genes$chromosome[i])
    for (pop in populations){
        output_file_name<-sprintf("plink_out/%s_%s", genes$gene[i], pop)
        keep_file_name<-sprintf("%s.keep", pop)
        args<-c("--vcf", file_path,
                "--keep", keep_file_name,
                "--chr", genes$chromosome[i],
                "--from-bp", genes$start[i],
                "--to-bp", genes$end[i],
                "--snps-only", "just-acgt",
                "--biallelic-only", "strict",
                "--make-bed",
                "--out", output_file_name)
        system2(cmd, args)
    }
}
```

## Linkage disequilibrium from real data
We are only keeping the SNP (just one nucleotide changes), but we haven't filtered the rare variants.

First we create the data with the coefficient r^2 for linkage disequilibrium, first excluding and then including rare variants.

```r
dir.create("linkage_disequilibrium", showWarnings = FALSE)
for (i in 1:3){
    for (pop in populations){
        input_file_name<-sprintf("plink_out/%s_%s", genes$gene[i], pop)
        output_file_name<-sprintf("linkage_disequilibrium/%s_%s_LD", genes$gene[i], pop)
        args<-c("--bfile", input_file_name,
                "--maf", "0.01", #only the variants with a frequency greater than 1%
                "--r2", "gz",
                "--ld-window-kb", "500",
                "--ld-window", "99999",
                "--ld-window-r2", "0", #keeps all the LD
                "--out", output_file_name
       )
        system2(cmd, args)
    }
}
```

```r
dir.create("linkage_disequilibrium_rare_variants", showWarnings = FALSE)
for (i in 1:3){
    for (pop in populations){
        input_file_name<-sprintf("plink_out/%s_%s", genes$gene[i], pop)
        output_file_name<-sprintf("linkage_disequilibrium_rare_variants/%s_%s_LD", genes$gene[i], pop)
        args<-c("--bfile", input_file_name,
                "--r2", "gz",
                "--ld-window-kb", "500",
                "--ld-window", "99999",
                "--ld-window-r2", "0", #keeps all the LD
                "--out", output_file_name
       )
        system2(cmd, args)
    }
}
```

Now we create the .csv files associated to those data

```r
dir.create("linkage_disequilibrium_csv", showWarnings = FALSE)
for (i in 1:3){
    for (pop in populations){
        input_file_name<-sprintf("linkage_disequilibrium/%s_%s_LD.ld.gz", genes$gene[i], pop)
        output_file_name<-sprintf("linkage_disequilibrium_csv/%s_%s_LD.csv", genes$gene[i], pop)
        table<-read_table(input_file_name)
        write.csv(table, output_file_name, row.names=FALSE)
    }
}
```

```r
dir.create("linkage_disequilibrium_rare_variants_csv", showWarnings = FALSE)
for (i in 1:3){
    for (pop in populations){
        input_file_name<-sprintf("linkage_disequilibrium_rare_variants/%s_%s_LD.ld.gz", genes$gene[i], pop)
        output_file_name<-sprintf("linkage_disequilibrium_rare_variants_csv/%s_%s_LD.csv", genes$gene[i], pop)
        table<-read_table(input_file_name)
        write.csv(table, output_file_name, row.names=FALSE)
    }
}
```

LD decay

```r
data_ld_decay<-data_ld %>%
    mutate(dist_bp=abs(BP_A - BP_B)) %>%
    mutate(dist_bin=as.integer(dist_bp/50)) %>%
    arrange(desc(dist_bin)) %>%
    group_by(dist_bin) %>%
    summarise(R2_mean=mean(R2))
head(data_ld_decay)
```

```r
start_a <- max(data_ld_decay$R2_mean)
start_l<-20
fit <- nls(
  R2_mean ~ a * exp(-dist_bin / lambda),
  data = data_ld_decay,
  algorithm="port",
  start = list(a = start_a, lambda = start_l),
  lower = c(c = 0, a = 0, lambda = 1e-6)
)
```

```r
base_to_number<-function(base){ #A=1, C=2, G=3, T=4
    stopifnot(base=="A" | base=="C" | base=="G" | base=="T")
    if(base=="A")
        return (1)
    if(base=="C")
        return (2)
    if(base=="G")
        return (3)
    if(base=="T")
        return (4)
}
```

## Markov chain generation of synthetic haplotypes

```r
data_abo_eur <- data_gnomad_rare_variants %>% # without rare variants
    filter(chromosome==9)
```

```r
nrow(data_abo_eur) #nothing changes
```

We want to create a vector with all the probabilities for the four different bases, only considering the positions for which we have a possible variant in the gnomAD dataset.

```r
probabilities<-tibble(
    `position`=0,
    `A`=0,
    `C`=0,
    `G`=0,
    `T`=0
)
row<-1
```

For each position we associate a probability to each possible variant, then the probability of the most common base (the reference) will be just 1-s, where s is the sum of the probabilities of all the other bases.

```r
for (pos in unique(data_abo_eur$position)){
    variants<-data_abo_eur %>%
        filter(position==pos)
    prob_vector<-rep(0, 5)
    for (i in 1:nrow(variants)){
        prob_vector[base_to_number(variants$alternate[i])+1]<-variants$allele_frequency[i]
    }
    prob_reference<-(1-sum(prob_vector))
    prob_vector[base_to_number(variants$reference[1])+1]<-prob_reference # 
    #reference[1] because we have at least one line and the reference base is always the same for the same position
    prob_vector[1]<-pos
    probabilities[row, 1:5] <- as.list(prob_vector)
    row<-row+1
}
```

Now we generate synthetic haplotypes using probabilities. Later we will add conditional probabilities.

```r
bases<-c("A", "C", "G", "T")
generate_base<-function(probabilities){
    weights<-probabilities[2:5]
    base<-sample(bases, size=1, prob=weights)
    return(base)
}
```

```r
synthetic_haplotypes_ABO_EUR <- tibble(
  `position`=probabilities$position
)
```

```r
for (i in 1:10000){
    haplotype<-rep(0, nrow(probabilities))
    for (j in 1:nrow(probabilities)){
        haplotype[j]<-generate_base(probabilities[j, ])
    }
    colname<-sprintf("haplotype_%d", i)
    synthetic_haplotypes_ABO_EUR[[colname]]<-haplotype
}
```

## Markov chain 2.2: using exponential model for LD decay

First we only try to condition the probability when we have alternate in the position before. When we have reference we keep the same probabilities.

```r
data_abo <- data_gnomad_rare_variants %>% # without rare variants
    filter(chromosome==9)
probabilities<-data_abo %>%
    group_by(position) %>%
    summarize(
        alternate_frequency=sum(allele_frequency),
        reference_frequency=1-alternate_frequency
    )
```

```r
head(probabilities)
```

```r
synthetic_haplotypes_ABO_EUR <- tibble(
  `position`=probabilities$position
)
```

```r
generate<-function(probabilities){
    weights<-probabilities[2:3]
    base<-sample(c(1, 0), size=1, prob=weights) # 0 means reference, 1 means alternate
    return(base)
}
```

Use the formula for the probability increase or decrease (we always suppose a positive correlation between alternates, since we only study r^2 and so we don't care whether it is correlation or anticorrelation).

When the probability of the mutation is greater than 1 we set it to 1.

```r
update_probability<-function(prob, allele_before, prob_allele_before, a, lambda){
    if(allele_before == 1){
        estimated_r<-sqrt(a)*exp(-(prob$position-prob_allele_before$position)/lambda)
        conditional<-sqrt((prob$reference_frequency*prob$alternate_frequency*prob_allele_before$reference_frequency)/prob_allele_before$alternate_frequency)*estimated_r
        prob$alternate_frequency<-prob$alternate_frequency + conditional
        if(prob$alternate_frequency>1)
            prob$alternate_frequency<-1
    }
    prob$reference_frequency<-1-prob$alternate_frequency
    return (prob)
}
```

```r
cppFunction('
    double update_probability_cpp(double p_alt, double p_ref, double position, 
    double p_alt_before, double p_ref_before, double position_before, double a, double lambda){
        double estimated_r, conditional;
        estimated_r=sqrt(a)*exp(-abs((position-position_before))/lambda);
        conditional=sqrt((p_ref*p_alt*p_ref_before)/p_alt_before)*estimated_r;
        p_alt=p_alt + conditional;
        if(p_alt>1.0){
            p_alt=1.0;
        }
        return p_alt;
    }
')
```

```r
cppFunction('
    IntegerMatrix generate_haplotypes_cpp(IntegerMatrix H, NumericVector prob_alternate, 
    NumericVector prob_reference, NumericVector position, double a, double lambda){
        for (int i=0; i<H.ncol(); ++i){
            NumericVector p_alt = clone(prob_alternate);
            NumericVector p_ref = clone(prob_reference);
            H(0, i) = static_cast<int>(R::runif(0.0, 1.0) < p_alt[0]);
            for (int j=1; j<H.nrow(); ++j){
                if(H(j-1, i)==1 && p_alt[j]<0.01){
                    double estimated_r, conditional;
                    estimated_r=sqrt(a)*exp(-abs((position[j]-position[j-1]))/lambda);
                    conditional=sqrt((p_ref[j]*p_alt[j]*p_ref[j-1])/p_alt[j-1])*estimated_r;
                    p_alt[j]=p_alt[j] + conditional;
                    if(p_alt[j]>1.0){
                        p_alt[j]=1.0;
                    }
                    p_ref[j]=1-p_alt[j];
                }
                 H(j, i) = static_cast<int>(R::runif(0.0, 1.0) < p_alt[j]);
            }
        }
        return H;
    }
')
```

```r
position<-probabilities$position
prob_reference<-probabilities$reference_frequency
prob_alternate<-probabilities$alternate_frequency
```

```r
update_probability<-function(p_alt, p_ref, position, p_alt_before, p_ref_before, position_before, a, lambda){
        estimated_r<-sqrt(a)*exp(-abs((position-position_before))/lambda)
        conditional<-sqrt((p_ref*p_alt*p_ref_before)/p_alt_before)*estimated_r
        p_alt<-p_alt + conditional
        if(p_alt>1)
            p_alt<-1
    return (p_alt)
}
```

```r
H <- matrix(0L, nrow = length(position), ncol = 10000L)
for (i in 1:ncol(H)){
    haplotype<-integer(nrow(H))
    p_alt<-prob_alternate
    p_ref<-prob_reference
    haplotype[1]<-as.integer(runif(1)<p_alt[1])
    for (j in 2:nrow(H)){
        if(haplotype[j-1]==1L & p_alt[j]<0.01){
            p_alt[j]<-update_probability(p_alt[j], p_ref[j], position[j], 
                p_alt[j-1], p_ref[j-1], position[j-1], a_ABO_EUR, lambda_ABO_EUR
            )
            p_ref[j]<-1-p_alt[j]
        }
        haplotype[j]<-as.integer(runif(1)<p_alt[j])
    }
    H[ ,i]<-haplotype
}
```

```r
linkage_coefficient<-function(synthetic_haplotypes, frequencies, index_A, index_B){
    allele_A<-as.vector(unlist(synthetic_haplotypes[index_A, ]))
    allele_B<-as.vector(unlist(synthetic_haplotypes[index_B, ]))
    both_alternate<-allele_A*allele_B #this vector is 1 when both A and B are alternate and 0 otherwise
    frequency_A_and_B<-sum(both_alternate)/length(both_alternate)
    #now use the formula for r^2
    denominator<-(frequencies$alternate_frequency_haplotypes[index_A])*(frequencies$reference_frequency_haplotypes[index_A])*(frequencies$alternate_frequency_haplotypes[index_B])*(frequencies$reference_frequency_haplotypes[index_B])
    numerator<-(frequency_A_and_B-((frequencies$alternate_frequency_haplotypes[index_A])*(frequencies$alternate_frequency_haplotypes[index_B])))^2
    r_2<-numerator/denominator
    return(r_2)
}
```
