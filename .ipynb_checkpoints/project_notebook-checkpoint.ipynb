## 1. Data transformation

We begin the analysis by uploading and transforming the data from the gnomAD dataset for the three genes HLA-DRB1, TP53 and ABO

```r
install.packages("MCMCpack") #useful later for the Dirichlet distribution
```

```r
library(MCMCpack)
library(tidyverse)
```

```r
data_hla<-read_csv(
    "gnomAD_HLA-DRB1.csv"
)
data_tp<-read_csv(
    "gnomAD_TP53.csv"
)
data_abo<-read_csv(
    "gnomAD_ABO.csv"
)
```

Since it is easier to have a single dataset, we merge the data for the three genes into a single dataset. When we will need to separate the data for the three genes, we will filter on the chromosome number (which is 6 for HLA_DRB1, 9 for ABO and 17 for TP53).

```r
data_gnomad<-bind_rows(data_hla, data_tp, data_abo)
stopifnot(nrow(data_gnomad)==nrow(data_tp) + nrow(data_abo) + nrow(data_hla))
```

The data are already in a tidy format. So, we only have to change the name of the columns and to add some columns for the allele frequencies for each population.

```r
data_gnomad<-data_gnomad %>%
    set_names(
        ~str_to_lower(.) %>%            #remove capital letters
        str_replace_all(" - ", "_") %>% 
        str_replace_all(" ", "_") %>%
        str_replace_all("-", "_") %>%
        str_replace_all("\\(", "") %>%
        str_replace_all("\\)", "") %>%
        str_replace_all("/african_american", "") #with african we mean both african and african-american
    )

    #allele_frequency=allele_count/allele_number for each population
data_gnomad<-data_gnomad %>%
    mutate(
        allele_frequency_african=allele_count_african/allele_number_african,
        allele_frequency_east_asian=allele_count_east_asian/allele_number_east_asian,
        allele_frequency_admixed_american=allele_count_admixed_american/allele_number_admixed_american,
        allele_frequency_ashkenazi_jewish=allele_count_ashkenazi_jewish/allele_number_ashkenazi_jewish,
        allele_frequency_european_finnish=allele_count_european_finnish/allele_number_european_finnish,
        allele_frequency_middle_eastern=allele_count_middle_eastern/allele_number_middle_eastern,
        allele_frequency_european_non_finnish=allele_count_european_non_finnish/allele_number_european_non_finnish,
        allele_frequency_amish=allele_count_amish/allele_number_amish,
        allele_frequency_south_asian=allele_count_south_asian/allele_number_south_asian,
        allele_frequency_remaining=allele_count_remaining/allele_number_remaining
        
    )
colnames(data_gnomad)
```

### 1.1 SNPs and other variants

Almost all the variants present in the gnomAD database are SNPs. However, we have also variants coming from other types of DNA mutations, such as for example indels (which means "insertion or deletion"). The issue with these mutations different from SNPs is that their representation in the dataset is ambigous: sometimes more variants represent the same mutation. This becomes clearer considering for example the variants of position 7674360 in the TP53 gene.

```r
data_gnomad %>%
    filter(position==7674360) %>%
    select(position, chromosome, reference, alternate, allele_number, allele_count, allele_frequency)
```

We have 12 possible variants for this position, but almost all of them represent the same mutation, which is the insertion or deletion of one or more T after C. We are not even sure that the number of T changes from one variant to the other: sometimes these differences depend only on the way the data have been collected.

Since our main interest is in SNPs, but we also want to keep some of the other variants, we filter the dataset according to a rule: for each position we keep all the SNPs and only the non-SNP variant with the highest frequency.

```r
nrow(data_gnomad)
```

```r
data_non_SNP_to_keep<-data_gnomad %>%
    filter(nchar(reference)!=1 | nchar(alternate)!=1) %>% #keep only the non-SNP
    group_by(position) %>%
    slice_max(allele_frequency, n=1)

elements_to_keep<-data_non_SNP_to_keep$gnomad_id #select the gnomad_id of the non-SNP we want to keep
data_gnomad<-data_gnomad %>%
    filter((nchar(reference)==1 & nchar(alternate)==1) | gnomad_id %in% elements_to_keep)
```

```r
nrow(data_gnomad)
```

As we can see, this operation only eliminates 132 rows, which is only 2.5% of the variants for the three genes.

This is the dataset we will use for the analysis in the next section. Sometimes, however, we will study separately common variants (variants with frequency >0.001) and rare variants (with frequency <0.001). So, we create two datasets, one containing only the common variants and the other only the rare variants.

```r
data_gnomad_common_variants<-data_gnomad %>%
    filter(allele_frequency>=0.01)
data_gnomad_rare_variants<-data_gnomad %>%
    filter(allele_frequency<0.01)
```

### 2.1 Distribution of variants across the genes

First, we study how the distance between variants are distributed across the three genes. To do so, we have to add a column to the dataset specifying the distance of a variant from the previous one.

```r
distance_vector<-rep(0, nrow(data_gnomad)) 
distance_vector[1]<-1234567 #the first element has no meaning, see later
for (i in 1:(nrow(data_gnomad)-1)){
    distance_vector[i+1]<-data_gnomad$position[i+1]-data_gnomad$position[i]
}
data_gnomad<-data_gnomad %>%
    mutate(dist_from_previous_variant=distance_vector)
unique(data_gnomad$dist_from_previous_variant)
```

The value 0 means that we have multiallelic variants, i.e. two or more possible variants for the same position. The biggest values in modulus (-24927934 and 125556827, and 1234567 added manually) mean that we are changing gene. So we have to remove them before plotting an histogram of the distances. We also have to remove the 0 for the first variant in the dataset (which means nothing, since we don't have any variation before).

```r
data_histogram_distance<-data_gnomad %>%
    select(dist_from_previous_variant, chromosome) %>%
    filter(abs(dist_from_previous_variant)<100000) %>% #remove the distance value for the initial position of each gene
    mutate(dist_cap=pmin(dist_from_previous_variant, 20)) #before was 20
    #distances higher than 20 bp have been collected in the last bin for better visualization
    
ggplot(data=data_histogram_distance, aes(x=dist_cap, fill=factor(chromosome))) +
    geom_histogram(bins=21) + #one bin for each value of the distance from 0 to 20
    labs(title = "Distance variants histogram",
       x = "Distance (bp)",
       y = "Count")
```

```r
nrow(data_histogram_distance %>% filter(dist_from_previous_variant==0))
```

The histogram shows that almost all the distances between consecutive variants are no bigger than 10 bp. This means that variants in all the three genes are very frequent, and we have only a limited number of long regions for which no variant has been observed. We can also notice that the bin 0 has a count of 1206, meaning that we have many multiallelic variants.

We can better visualize the same data using a boxplot: it clearly shows that more than 75% of the distances are between 0 and 2 bp.

```r
ggplot(data=data_histogram_distance, aes(x=factor(chromosome), y=dist_cap, fill=factor(chromosome))) +
    geom_boxplot() +
    stat_boxplot( geom = "errorbar" ) +
    labs(title = "Distance variants boxplot",
       x = "Chromosome",
       y = "Distance (bp)")
```

If we remove the rare variants, we can see that almost all the common variants are located in the gene HLA-DRB1. This is interesting, since this gene is the one with the smallest number of variants between the three genes.

```r
#this is the same code as before but applied on the dataset with only common variants

distance_vector<-rep(0, nrow(data_gnomad_common_variants)) 
distance_vector[1]<-1234567
for (i in 1:(nrow(data_gnomad_common_variants)-1)){
    distance_vector[i+1]<-data_gnomad_common_variants$position[i+1]-data_gnomad_common_variants$position[i]
}

data_gnomad_common_variants<-data_gnomad_common_variants %>%
    mutate(dist_from_previous_variant=distance_vector)

data_histogram_distance_common_variants<-data_gnomad_common_variants %>%
    select(dist_from_previous_variant, chromosome) %>%
    filter(abs(dist_from_previous_variant)<100000) %>% #remove the distance value for the initial position of each gene
    mutate(dist_cap=pmin(dist_from_previous_variant, 50))
    #in this case we keep distances up to 50 since we have more higher values

ggplot(data=data_histogram_distance_common_variants, aes(x=dist_cap, fill=factor(chromosome))) +
    geom_histogram(bins=51) + #one bin for each value of the distance from 0 to 50
    labs(title = "Distance variants histogram",
       x = "Distance (bp)",
       y = "Count")
```

```r
#comparison between the total number of variants and the number of common variants in each gene
data_gnomad %>%
    group_by(chromosome) %>%
    summarize(count=n())

data_gnomad_common_variants %>%
    group_by(chromosome) %>%
    summarize(count=n())
```

All these plot do not give any information about how the variants are distributed in the gene. So, we divide each gene into subregions of 1000 bp, and we count the number of variants for each subregion.

```r
data_variants_distribution<-data_gnomad %>%
    mutate(dist_from_previous_variant=ifelse(abs(dist_from_previous_variant)>100000, 0, dist_from_previous_variant)) %>%
    mutate(rescaled_position=cumsum(dist_from_previous_variant)) %>%
    mutate(rescaled_position=rescaled_position+position[1]) %>%
    #these transformations are done just to have the three genes on the same histogram, one next to the other
    mutate(window=as.integer(rescaled_position/1000)) %>%
    mutate(window=ifelse(chromosome==17, window+1, window)) %>% #separated bins for each gene
    mutate(window=ifelse(chromosome==9, window+2, window))

ggplot(data=data_variants_distribution, aes(x=window, fill=factor(chromosome))) +
    geom_histogram(bins=80) + 
    labs(title = "Distribution of variants",
       x = "Subregion",
       y = "Count") 
```

The distributions of variants for the three genes have been incorporated in the same histogram just for better visualization, but they must be considered separately. We can notice that the variants are not uniformly distributed along the genes: we have small regions where many variants are concentrated together with larger regions without any variant. This could suggest the existance of a link between variants: a mutation in a certain position could induce mutations in nearby positions.

If we now consider the mean value and standard deviation of frequency in each window, we can see that we have the highest variant frequencies in the HLA-DRB1 gene. This is due to the highest number of common variants in this gene, 

```r
data_mean_frequency<-data_variants_distribution %>%
    group_by(window) %>%
    summarize(mean_fr=median(allele_frequency), std_fr=sd(allele_frequency), chromosome=max(chromosome))
ggplot(data=data_mean_frequency, aes(x=window, y=mean_fr, color=as.factor(chromosome))) +
    geom_point() 
    #geom_errorbar(aes(ymin=mean_fr-std_fr, ymax=mean_fr+std_fr))
```

All the Pearson's correlation coefficients between the couples of populations are very high (the lowest one is 0.85): this means that we have a strong linear relation between allele frequencies for different populations. ### 2.3 Correlation between allele frequencies at different positions
# COMMENTA IL CODICE (IN GENERALE RICONTROLLA TUTTO IL CODICE E COMMENTALO)

As we said in the Introduction, to study linkage disequilibrium across variants we need data about the co-occurrence of variants in the same genotype, and these data are not present in the gnomAD dataset. The only way to get a rough approximation of linkage disequilibrium is to treat the frequencies of a variant across different populations as a dataset, and then calculate the correlations between frequencies at different positions. so, while in the last paragraph we studied the correlations between columns of the gnomAD dataset, here we analyze the correlations between rows.

We want to do a separate analysis for the three genes. So, we write a function that takes a gnomAD dataset and computes the correlation coefficient between all possible couples of variants, returning a dataset containing the correlation coefficient in function of the distance between variant positions.

```r
frequency_correlations<-function(dataset, method, chromosome){
    data_correlation_variants<-dataset %>%
        select(position, variant_number, allele_frequency_african,
            allele_frequency_east_asian, allele_frequency_admixed_american, allele_frequency_ashkenazi_jewish,
            allele_frequency_european_finnish, allele_frequency_middle_eastern, allele_frequency_european_non_finnish,
            allele_frequency_amish, allele_frequency_south_asian, allele_frequency_remaining
        )
    data_correlation_transposed <- data_correlation_variants %>%
    mutate(position_variants=paste0(as.character(position), "_", as.character(variant_number))) %>%
    column_to_rownames("position_variants") %>%  # position → rownames
    t() %>%                              # trasposizione
    as.data.frame() %>%
    rownames_to_column("population") %>%
    select(!population)

    #index<-which(data_correlation_transposed==0, arr.ind = TRUE)
    cor_matrix_populations<-cor(data_correlation_transposed, method=method, use="pairwise.complete.obs")
    position_1<-rownames(cor_matrix_populations)
    position_2<-colnames(cor_matrix_populations)

    position_1<-sub("_.", "", position_1) #removes the variant number from the position
    position_2<-sub("_.", "", position_2)

    dist_mat <- outer(as.numeric(position_1), as.numeric(position_2), "-")
    chrom_vec<-rep(chromosome, length(as.vector(dist_mat)))
    dataset_r<-tibble(
    `distance` = as.vector(dist_mat),
    `R` = as.vector(cor_matrix_populations),
    `chromosome`= chrom_vec
    )
    return(dataset_r)
}
```

Then we calculate separately the correlation coefficients for the three genes and merge them into a single dataset.

(We have however to add the variant number immediately)

```r
correlations_HLA_DRB1<-frequency_correlations(data_gnomad %>% filter(chromosome==6), "spearman", 6)
correlations_TP53<-frequency_correlations(data_gnomad %>% filter(chromosome==17), "spearman", 17)
correlations_ABO<-frequency_correlations(data_gnomad %>% filter(chromosome==9), "spearman", 9)
```

```r
head(correlations_TP53)
```

```r
correlations_all_genes<-bind_rows(correlations_HLA_DRB1, correlations_TP53, correlations_ABO)

correlations_all_genes<-correlations_all_genes %>%
    filter(distance>0) %>%
    mutate(dist_bin=as.integer(distance/50)) %>%
    arrange(desc(dist_bin)) %>%
    group_by(dist_bin, chromosome) %>%
    summarise(R_mean=mean(R))

ggplot(data=correlations_all_genes, aes(x=dist_bin, y=R_mean)) +
geom_point(aes(color=factor(chromosome)))
```

## 3. Generation of synthetic haplotypes

In this section we use the allele frequency data from gnomAD to generate synthetic haplotypes for the three genes. We only consider the positions with at least one variant (i.e. the positions that are present in the dataset), since we don't have any information about the other positions.

To simplify the simulation, we represent a position in which we have a variant as a 1 and a position in which we don't have a variant as a 0, without considering which base or group of bases we have at a certain position. However, we have to consider the existance of more variants at the same position

```r
# count the maximum number of variants we can have at a certain position
count_different_variants<-data_gnomad %>%
    group_by(position) %>%
    summarize(count=n())
    
unique(count_different_variants$count)
```

Since we can have up to 4 variants for the same position, we assign a progressive number to each variant. This number is 1 for the first variant of the position, 2 for the second variant and so on up to 4.

```r
data_gnomad<-data_gnomad %>%
    mutate(variant_number=1L) %>% #L means integer in R
    mutate(variant_number=ifelse(dist_from_previous_variant==0, lag(variant_number)+1L, 1L)) %>%
    mutate(variant_number=ifelse(dist_from_previous_variant==0, lag(variant_number)+1L, 1L)) %>%
    mutate(variant_number=ifelse(dist_from_previous_variant==0, lag(variant_number)+1L, 1L))
#gives an index for each variant at each position. It is always 1 unless we have more than one variant
#at a certain position. We repeat the 'mutate' 4 times because each time it modifies 'variant_number'
#according to the former value of 'variant_number'
```

### 3.1 Independent sampling

We generate the haplotypes using the simplest possible model: independent sampling. For each position we have a probability $p_i$ ($i=1,..,K$), where K is the number of variants in that position) to have the $\textit{i-th}$ alternate allele and a probability $p_r=1-\sum_{i=1}^Kp_i$ to have the reference allele. These probabilities are independent from the presence or absence of other variants in the same haplotype.

In practice, for each position we sample a value according to a multinomial Bernoulli distribution with parameters $(p_1,...,p_K,p_r)$. Therefore the parameters follow the conjugated prior distribution of the multinomial Bernoulli distribution, which is the Dirichlet distribution (a multinomial Beta distribution). 

So, for a certain position with N variants we start from a prior Dirichlet distribution with parameters $(\alpha_1,...,\alpha_{K}, \alpha_r)=(1,..,1), then we update it following the number of times a certain variant has been observed (given by the column "allele_count" in the gnomAD dataset):

$$
\begin{equation*}
    \alpha_i \rightarrow \alpha_i + AC_i
\end{equation*}\\
\begin{equation*}
    \alpha_r \rightarrow \alpha_r + AN - \sum_{i=1}^KAC_i
\end{equation*}
$$

where $AC_i$ is the "allele_count" for the variant $i$ and $AN$ is the total number of alleles observed (the column "allele_number" in the gnomAD dataset). The total number of alleles observed is different between different variants in the same position, but the difference is negligible compared to the number of observations, so it is not important which one do we choose. The choice of the prior is not very important, too, since we have a huge number of observed alleles (order of $10^5-10^6$), so the choice of a value of 1 for all the prior hyperparameters does not really affect the posterior distribution.

From this posterior distribution we sample the parameters of the multinomial Bernoulli distribution we will use for haplotypes generation.

Now we define a function that takes a gnomAD dataset and generates a matrix with N haplotypes.

```r
generate_independent_haplotypes<-function(dataset, N){
    haplotypes_matrix <- matrix(0, nrow = length(unique(dataset$position)), ncol = N)
    row<-1
    for (pos in unique(dataset$position)){
        variants<-dataset %>%
            filter(position==pos) #the number of rows of variants is the number of variants for the position
        parameters_vector<-rep(1, nrow(variants)+1) # +1 to include also variant 0 which is the reference

        for (i in 1:nrow(variants)){
            parameters_vector[i+1]<-1+variants$allele_count[i]
        }
        parameters_vector[1]<-1+max(variants$allele_number)-sum(variants$allele_count)

        #we have updated the parameters and obtained the posterior distribution of the probabilities
        #now we generate the probabilities of the variants

        prob_vector<-rdirichlet(n=1, alpha=parameters_vector)

        #and finally we generate for all the haplotypes the value they take in this position

        haplotypes_matrix_row<-sample.int(n=length(prob_vector), size=N, prob=prob_vector, replace=TRUE)-1
        haplotypes_matrix[row, ]<-haplotypes_matrix_row
        row<-row+1
    }
    colnames(haplotypes_matrix) <- sprintf("haplotype_%d", seq_len(N)) #give names to the matrix columns

    #now we transform the matrix into a dataset, with position and haplotypes as columns

    synthetic_haplotypes <- tibble(`position`=unique(dataset$position))
    synthetic_haplotypes <- bind_cols(
        synthetic_haplotypes,
        as_tibble(haplotypes_matrix)
    )
    return(synthetic_haplotypes)
}
```

In this section we only define a function for the generation. We will generate haplotypes only in the next sections, choosing the dataset according to the type of analysis we want to perform (for example including only common variants or only a certain population).

We also create a function that gives the frequencies of the variants in the generated haplotypes.

```r
get_frequencies<-function(dataset, synthetic_haplotypes){
    frequencies<-rep(0, nrow(dataset))
    row_numbers<-match(dataset$position, synthetic_haplotypes$position)
    #match the positions for the synthetic haplotypes with the positions in the gnomAD dataset, useful since
    #in the gnomAD dataset we have sometimes more than one row for position
    
    haplotypes_matrix<-as.matrix(synthetic_haplotypes[,-1])
    rownames(haplotypes_matrix)<-synthetic_haplotypes$position
    #convert the dataset to a matrix, which is much faster in the next part of the code

    for (i in 1:nrow(dataset)){
        variant_number<-dataset$variant_number[i]
        frequencies[i]<-sum(as.integer(haplotypes_matrix[row_numbers[i], ]==variant_number))/length(haplotypes_matrix[row_numbers[i], ])
    }
    return(frequencies)
}
```

## 4. Variant structure in synthetic data

In this section we compare real data and synthetic data. Since we considered only the positions in the three genes where we have variants in the gnomAD dataset, it is clear that the spatial distribution of variants will be the same for real and simulated data. For this reason we focus only on frequencies of variants.

```r
synthetic_haplotypes<-generate_independent_haplotypes(data_gnomad, 10000)
```

```r
synthetic_haplotypes_HLA_DRB1<-generate_independent_haplotypes(data_gnomad %>% filter(chromosome==6), 10000)
synthetic_haplotypes_TP53<-generate_independent_haplotypes(data_gnomad %>% filter(chromosome==6), 10000)
synthetic_haplotypes_ABO<-generate_independent_haplotypes(data_gnomad %>% filter(chromosome==6), 10000)
```

### 4.1 Frequencies from real and simulated data

First of all we want to be sure that the model correctly reproduces the frequencies of variants in real data. The easiest way to verify it is to do a linear regression to study the relationship between frequencies in real and simulated data.

```r
linear_relation_real_and_simulated<-lm(synthetic_frequency ~ allele_frequency, data=data_gnomad_very_rare)
summary(linear_relation_real_and_simulated)
```

```r
ggplot(data = data_gnomad_very_rare, aes(x=allele_frequency, y=synthetic_frequency)) +
geom_point( ) +
geom_smooth( method = "lm") +
labs(x = "Real data frequency",
       y = "Synthetic haplotypes frequency")
```

From the linear regression model we obtain exactly what we expect: a regression line with a slope of 1 and an intercept of 0. This means that the frequencies of variants are the same in real and simulated data.

We repeat the same operation, but keeping only frequencies lower than $10^{-4}$. As we can see the linear relation is no longer true, since we have many variants with small frequency that never appear. So we need to generate a bigger number of haplotypes.

### 4.2 Haplotype diversity

We want to see if our model creates some haplotypes that are more common than the others. To do so, we assign to each haplotype a number that characterizes that haplotype. This number is obtained by multiplying the variant number at a certain position of the haplotype by the position number, and then summing all these numbers.

```r
synthetic_haplotypes_diversity_common_variants<-synthetic_haplotypes_common_variants %>%
  mutate(across(-position, ~ .x * position))
#multiplies all variants in the haplotypes by the corresponding position

synthetic_haplotypes_diversity_common_variants <- synthetic_haplotypes_diversity_common_variants %>%
  column_to_rownames("position") %>%  
  t() %>%                              # exchange columns and rows
  as.data.frame() %>%
  rownames_to_column("haplotype") %>%
  mutate(number = rowSums(across(-haplotype), na.rm = TRUE)) %>%
  arrange(number) %>%
  mutate(index=row_number())
```

```r
ggplot(data=synthetic_haplotypes_diversity_common_variants, aes(x=index))+
    geom_histogram(bins=10000)
```

The histogram shows exactly what we expect to see: since the variant at each position is sampled independently, all the haplotypes are different and we don't have a more frequent pattern.

We drew the histogram considering only common variants, but clearly the result wouldn't change by adding rare variants, since we would only add more variability to the haplotypes.

### 4.3 Number of variants per position

We want to see how the variants are distributed along the genes. So, we divide again the genes into 1000-bp windows, and count for each haplotype how many variants we have in a window. Then we average over all the haplotypes and see the result.

In this case we only need to know if we have or not a variant in a certain position, without caring about which variant it is (if we have more than one variant for a position).

```r
data_rescaled<-data_gnomad_common_variants %>%
    mutate(dist_from_previous_variant=ifelse(abs(dist_from_previous_variant)>100000, 0, dist_from_previous_variant)) %>%
    mutate(rescaled_position=cumsum(dist_from_previous_variant)) %>%
    mutate(rescaled_position=rescaled_position+position[1]) %>%
    #these transformations are done just to have the three genes on the same histogram, one next to the other
    mutate(window=as.integer(rescaled_position/500)) %>%
    mutate(window=ifelse(chromosome==17, window+1, window)) %>% #separated bins for each gene
    mutate(window=ifelse(chromosome==9, window+2, window)) %>%
    group_by(position) %>%
    slice(1) %>% #take one value for each position, we don't want duplicates for the join in the next cell
    ungroup() %>%
    select(position, window, chromosome)
```

```r
data_window_distribution_common_variants<-synthetic_haplotypes_common_variants %>%
    left_join(data_rescaled, by="position") %>%
    mutate(across(-c("position", "window", "chromosome"), ~ as.integer(.x != 0))) %>%
    group_by(window) %>%
    summarize(across(-c("position", "chromosome"), ~ sum(.x)), chromosome=max(chromosome)) %>%
    mutate(mean_count=rowMeans(across(starts_with("haplotype")))) %>%
    mutate(std=apply(across(starts_with("haplotype")), 1, sd))
```

```r
head(data_window_distribution_common_variants)
```

```r
ggplot(data=data_window_distribution_common_variants, aes(x=window, y=mean_count, color=factor(chromosome))) +
    geom_point() +
    geom_errorbar(aes(ymin=mean_count-std, ymax=mean_count+std))
```

```r
r_2_matrix<-cor(transposed_haplotypes, method="pearson", use = "pairwise.complete.obs")
```

```r
position_1<-rownames(r_2_matrix)
position_2<-colnames(r_2_matrix)

dist_mat <- outer(as.numeric(position_1), as.numeric(position_2), "-")

dataset_r_2<-tibble(
  `distance` = as.vector(dist_mat),
  `R2` = as.vector(r_2_matrix^2)
)

dataset_r_2<-dataset_r_2 %>%
    filter(distance>0)%>%
    mutate(dist_bin=as.integer(distance/50)) %>%
    arrange(desc(dist_bin)) %>%
    group_by(dist_bin) %>%
    summarise(R2_mean=mean(R2))
```

```r
ggplot(data=dataset_r_2, aes(x=dist_bin, y=R2_mean)) +
    geom_point()
```
